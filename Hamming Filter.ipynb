```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_detectors = 512  # Number of detectors (can be any number)
freqs = np.fft.fftfreq(n_detectors)
freq_cutoff = 0.5
Hamming = (np.abs(freqs))*(0.54 + 0.46*np.cos(2*np.pi*freqs/freq_cutoff))

#To shift the baseline frequencies to the centre point (0)
shifted_freqs = np.fft.fftshift(freqs)
shifted_Hamming = np.fft.fftshift(Hamming)

# Plot
plt.figure(figsize=(15, 8))
plt.plot(shifted_freqs, shifted_Hamming, 'g-', linewidth=2)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Filter Magnitude', fontsize=12)
plt.title('Hamming Filter', fontsize=16)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlim(-0.25, 0.25)
plt.ylim(-0.05, 0.15)
plt.show()

```
